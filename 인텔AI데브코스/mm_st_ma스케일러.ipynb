```python
from __future__ import print_function
import os
data_path = ['C:\\data']
import pandas as pd
# Import the data using the file path
filepath = os.sep.join(data_path + ['Orange_Telecom_Churn_Data.csv'])
data = pd.read_csv(filepath)
data.head().T
```

```python
# Remove extraneous columns
data.drop(['state', 'area_code', 'phone_number'], axis=1, inplace=True)
data.columns
from sklearn.preprocessing import LabelBinarizer
#tf를 01로 고치기
lb = LabelBinarizer()
for col in ['intl_plan', 'voice_mail_plan', 'churned']:
    data[col] = lb.fit_transform(data[col])
data.head()
```

```python
data_1 = data
data_2 = data
data_3 = data
```

```python
# Mute the sklearn warning
import warnings
warnings.filterwarnings('ignore', module='sklearn')
#sklearn은 기본적으로 넘파이형으로 받음
#최대 최소 1,0으로 바꾸기
from sklearn.preprocessing import MinMaxScaler

msc = MinMaxScaler()

data_msc = pd.DataFrame(msc.fit_transform(data_1),  # this is an np.array, not a dataframe.
                    columns=data_1.columns)
data_msc.head()
```

```python
#StandardScaler
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
#sklearn은 기본적으로 넘파이형으로 받음

data_ss = pd.DataFrame(ss.fit_transform(data_2),  # this is an np.array, not a dataframe.
                    columns=data_2.columns)
data_ss.head()
```

```python
# Mute the sklearn warning
from sklearn.preprocessing import MaxAbsScaler
mas = MaxAbsScaler()

data_mas = pd.DataFrame(mas.fit_transform(data_3),  # this is an np.array, not a dataframe.
                    columns=data_3.columns)
data_mas.head()
```

```python
#데이터 분리하기
#Get a list of all the columns that don't contain the label
x_cols1 = [x for x in data_msc.columns if x != 'churned']
x_cols2 = [x for x in data_ss.columns if x != 'churned']
x_cols3 = [x for x in data_mas.columns if x != 'churned']

# Split the data into two dataframes
#x데이터에 churned를 빼고 다시 넣어줌
X_data1 = data_msc[x_cols1]
y_data1 = data_msc['churned']
X_data2 = data_ss[x_cols2]
y_data2 = data_ss['churned']
X_data3 = data_mas[x_cols3]
y_data3 = data_mas['churned']

# # alternatively:
# X_data = data.copy()
# y_data = X_data.pop('churned')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn1 = KNeighborsClassifier(n_neighbors=3)
knn1 = knn1.fit(X_data1, y_data1)
y_pred1 = knn1.predict(X_data1)
y_pred1
```

```python
#정확도 구하기식
# Function to calculate the % of values that were correctly predicted
# 비교해서 정확도 재는 알고리즘 만들고 출력하기
def accuracy(real, predict):
    #같을때는 더해줌 
    return sum(real == predict) / float(real.shape[0])
```

```python
print(accuracy(y_data1, y_pred1))
```

```python
knn2 = KNeighborsClassifier(n_neighbors=3)
knn2 = knn2.fit(X_data2, y_data1)#y_data2는(원래 T,F이지만) 0과 1로만 나타내어 지지 않음!!!
y_pred2 = knn2.predict(X_data2)
y_pred2
```

```python
print(accuracy(y_data1, y_pred2))
```

```python
knn3 = KNeighborsClassifier(n_neighbors=3)
knn3 = knn3.fit(X_data3, y_data3)
y_pred3 = knn3.predict(X_data3)
y_pred3
```

```python
print(accuracy(y_data3, y_pred3))
```
